Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot. 


import numpy as np


np.random.seed(3)


x = 3 + np.random.normal(0, 2, 50)


y = 3 + np.random.normal(0, 2, len(x))


Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


import numpy as np


company = np.array(["Apple", "Microsoft", "Google", "AMD"])


profit = np.array([3000, 8000, 1000, 10000])

```python
#Ans:--
    # A bar plot  is a graphical representation of categorical data using rectangular bars with heights or lengths proportional to the values they represent.
    # It is commonly used to compare categories of data and make comparisons.

import numpy as np 
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft" ,"Google" , "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
x = company
y = profit
plt.bar(x,y)
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


box1 = np.random.normal(100, 10, 200)


box2 = np.random.normal(90, 20, 200)

```python
#Ans:--
    # a box plot is a standardized way of displaying the distribution of a continuous variable based on its five-number summary of data points
    # it is used to indicate a distribution is skewed or not and if there are potential unusual observations (outliers) present in the data set
    
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.show()
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.show()
```
